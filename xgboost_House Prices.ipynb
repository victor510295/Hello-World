```python
import os
import pandas as pd

os.chdir('C:\\Users\\Justin\\Desktop\\DM\\House Prices')
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train['TypeData'] = pd.Series(1, index=train.index)
test['TypeData']  = pd.Series(2,  index=test.index)
Bind = train.append(test, ignore_index=False, sort=False)
Bind.head(6)
```

```python
DummyStr = Bind.columns[Bind.dtypes == object]

BindDummy = Bind

for i in DummyStr:
  print('Now is ',i)
  Dummy = pd.get_dummies(Bind.loc[:, i], prefix=i)
    
  # DummyStrTitanic.head(6)
  # AttributeError: 'list' object has no attribute 'head'
  print(Dummy[1:6])

  BindDummy = pd.concat( [BindDummy, Dummy], axis = 1 )
```

```python
BindDummy = BindDummy.drop(DummyStr, axis = 1 )
BindDummy.head(6)
```

```python
XTrain = BindDummy[BindDummy.TypeData == 1].drop(['SalePrice'],axis=1)
YTrain = BindDummy[BindDummy.TypeData == 1].loc[:,'SalePrice']
```

```python
import xgboost as xgb

xgbc = xgb.XGBRegressor()
xgbc.fit(XTrain, YTrain)
```

```python
XTest = BindDummy[BindDummy.TypeData == 2].drop(['SalePrice'],axis=1)
YTest = xgbc.predict(XTest)
```

```python
XTest.head(6)
```

```python
XTest2Df = pd.DataFrame(data=XTest.Id, columns=['Id'])
YTest2Df = pd.DataFrame(data=YTest,    columns=['SalePrice'])
Result = pd.concat([XTest2Df, YTest2Df], axis=1)
Result.head(6)
```

```python
import time
NameFile = 'Submission_'+time.strftime("%y%m%d_%H%M%S", time.localtime())+'.csv'
print(NameFile)
Result.to_csv(NameFile, index=False)
```
